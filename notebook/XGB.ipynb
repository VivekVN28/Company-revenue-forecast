```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

colors = sns.color_palette("husl", 8)
```

```python
df=pd.read_csv('data/VISASMIDSA.csv')
df.head()
```

```python
df1=pd.read_csv('data/CPIAUCSL.csv')
df1.head()
```

```python
df1=df1.set_index('observation_date')
df1.index=pd.to_datetime(df1.index)
df1.plot()
```

```python

df=df.set_index('observation_date')
df.index=pd.to_datetime(df.index)
df.plot()
plt.show()
```

```python
df_plot = (df - df.mean()) / df.std()
df_plot1 = (df1 - df1.mean()) / df1.std()
df_plot1.plot(figsize=(10,5))
df_plot.rolling(5).mean().plot(figsize=(10,5))
```

```python
train=df.loc[df.index<'2022-01-01']
test=df.loc[df.index>='2022-01-01']

fig,ax=plt.subplots(figsize=(10,6))
ax.plot(train.index,train['VISASMIDSA'],label='Train',color=colors[0])
ax.plot(test.index,test['VISASMIDSA'],label='Test',color=colors[1])
ax.legend()
```
